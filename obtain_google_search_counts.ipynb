```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
pd.set_option('display.max_columns', None)
import time
```

```python
df = pd.read_csv('COFFEE_data.csv')
df.head()
```

```python
def get_results(city_name,name):
    search_query = [city_name[0] + ' ' + name[0]]

    cont=1
    count=0
    while cont:
        count+=1
        re = requests.get('https://www.google.com/search',
                          params={'q':search_query})
        soup = BeautifulSoup(re.text, 'lxml')
        response = soup.find('div', {'id': 'resultStats'})
        if not(response==None): # If result is not None
            cont=0
            try: # Catch issue where there are 0 search results found
                return int(response.text.replace(',', '').split()[1])
            except:
                return int(response.text.replace(',', '').split()[0])
        if count==9: # After trying multiple times just return
            cont=0
            print('Google not responding')
            return 0
```

```python
sleep_time=0.2

unique_names = (df['City']+ ',' +df['Store_Name'])

num_results={}
count=0
print('Beginning searches for Google Search Results...')
for item in unique_names:
    count+=1
    # Split by empty space
    city_name=['+\'' + item.split(',')[0]]
    shop_name=['+\''+ item.split(',')[1] +'\'']
    if (count%100): # Print every 100 results
        print('Examining item %i out of %i' %(count,len(unique_names)))
    num_results[item] = get_results(city_name,shop_name)

    # Pause before conducting next search
    time.sleep(sleep_time)
print('All Google Search Results have been calculated!')
```

```python
city_list=[]
store_list=[]
search_counts=[]
for item in unique_names:
    city_list.append(item.split(',')[0])
    store_list.append(item.split(',')[1])
    search_counts.append(num_results[item])
    
search_df=pd.DataFrame()
search_df['City'] = city_list
search_df['Store Name'] = store_list
search_df['Search Counts'] = search_counts

print(search_df.head())

search_df.to_csv('search_counts.csv')
```
